```python
#generate hypercube

def graph_cart_product(A,B=None):
    if B is None: B=A.copy()
    
    n,m = A.shape[0],B.shape[0]
    return np.kron(A,np.eye(m)) + np.kron(np.eye(n),B)
```

```python
#generate hypercube

def strong_product(A,B=None):
    if B is None: B=A.copy()
    
    n,m = A.shape[0],B.shape[0]
    return np.kron(A+ np.eye(n),np.eye(m)+B) - np.eye(n+m)
```

```python
def eig_sort(M):
    eigenval,v = np.linalg.eig(M)
    v = v[:,np.argsort((eigenval))[::-1]]
    eigenval = eigenval[np.argsort((eigenval))[::-1]]
    return eigenval,v
```

```python
#cartesian product of two path graphs
def gen_path_graph(n):
    edges = [[i,i+1] for i in range(n-1)]
    edges = np.concatenate([edges,[np.flipud(i) for i in edges]])
    A = np.zeros([n,n])
    D = np.eye(n)
    for e in edges:
        A[e[0],e[1]] = 1
    D = np.eye(n)*A.sum(axis=1)
    L = D - A
    return A,D,L


def gen_path_plus(n):
    edges = [[i,i+1] for i in range(n-1)]
    edges = np.concatenate([edges,[np.flipud(i) for i in edges]])
    #edges = np.concatenate([edges,[[n-1,3],[3,n-1]]])
    edges = np.concatenate([edges,[[2,4],[4,2]]])
    A = np.zeros([n,n])
    D = np.eye(n)
    for e in edges:
        A[e[0],e[1]] = 1
    D = np.eye(n)*A.sum(axis=1)
    L = D - A
    return A,D,L

def gen_cycle_graph(n):
    edges = [[i,(i+1)%(n)] for i in range(n)]
    edges = np.concatenate([edges,[np.flipud(i) for i in edges]])
    A = np.zeros([n,n])
    D = np.eye(n)
    for e in edges:
        A[e[0],e[1]] = 1
    D = np.eye(n)*A.sum(axis=1)
    L = D - A
    return A,D,L

def gen_star_graph():
    return None
def gen_tree_graph(depth=3,max_b=3):
    
    edges = []
    ctr = 0
    for d in range(depth):
        n_branch = np.random.randint(0,max_b)
        
    
def simple_tree(n=8):
    edges = [[0,1],[1,2],[1,3],[2,4],[2,5],[3,6],[3,7]]
    edges = np.concatenate([edges,[np.flipud(i) for i in edges]])
    A = np.zeros([n,n])
    D = np.eye(n)
    for e in edges:
        A[e[0],e[1]] = 1
    D = np.eye(n)*A.sum(axis=1)
    L = D - A
    return A,D,L

def gen_star_graph(n=0):
    n=5
    edges = [[0,1],[0,2],[0,3],[0,4]]
    edges = np.concatenate([edges,[np.flipud(i) for i in edges]])
    A = np.zeros([n,n])
    D = np.eye(n)
    for e in edges:
        A[e[0],e[1]] = 1
    D = np.eye(n)*A.sum(axis=1)
    L = D - A
    return A,D,L

def gen_random_graph(gtype='ER',n=5,m=4,p=.3):
    if gtype=='ER':
        g = nx.random_graphs.erdos_renyi_graph(n,.7)
    elif gtype=='BB':
        g = nx.generators.barbell_graph(n,2)
    elif gtype=='FC':
        g = nx.complete_graph(n)
    elif gtype=='EMPTY':
        g = nx.empty_graph(n=n)
        g.add_edges_from([[1,0],[0,1]])
    elif gtype=='gnm':
        done = False
        while not done:
            g = nx.random_graphs.gnm_random_graph(5,4)
            done = nx.is_connected(g)
            
    elif gtype=='WS':
        g = nx.connected_watts_strogatz_graph(n,m,p=.3)
    elif gtype=='ROOM':
        g = nx.grid_2d_graph(n,m)
        
    elif gtype=='BI':
        done = False
        while not done:

            g = nx.bipartite.random_graph(n,m,p)
            done = nx.is_connected(g)

    A = nx.to_numpy_matrix(g)
    n = A.shape[0]
    D = np.eye(n)*A.sum(axis=1)
    L = D - A

    return A,D,L


def gen_outerplanar():
    n = 8
    edges = [[0,1],[1,2],[2,3],[2,0],[3,4],[3,0],[4,5],[5,6],[5,3],[6,7],[6,3],[7,3],[7,0]]
    edges = np.concatenate([edges,[np.flipud(i) for i in edges]])
    A = np.zeros([n,n])
    D = np.eye(n)
    for e in edges:
        A[e[0],e[1]] = 1
    D = np.eye(n)*A.sum(axis=1)
    L = D - A
    return A,D,L
```

But inside the network what you are doing is factoring it into single or mixtures of updates of different components of the network.

```python
from torch import nn

class Network(nn.Module):
    def __init__(self):
        super().__init__()
        
        # Inputs to hidden layer linear transformation
        self.embed_action = nn.Linear(4,8)
        self.embed_action2 = nn.Linear(8,8)
        # Output layer, 10 units - one for each digit
        #self.output = nn.Linear(64, 5)
        
        
        # Define sigmoid activation and softmax output 
        self.nlin = nn.Tanh()
        self.softmax = nn.Softmax()
        
        self.compress1 = nn.Linear(144,8)
        self.uncompress = nn.Linear(8,144)
        
    def forward(self,action,x2):
        
        #this is embedded action
        ae = self.nlin(self.embed_action(action))
        ae2 = self.nlin(self.embed_action2(ae))
        
        #first compress the current state representation
        x2_ = self.nlin(self.compress1(x2))  #this represents the current place
        
        #print(x2_.shape,ae2.shape)
        #update current place with action
        x2_u = x2_ * ae2#tf.dot(tf.transpose(x2_,0,1),ae2)
        
        x_prev = self.uncompress(x2_)
        x_next = self.uncompress(x2_u)
        
        
        
        
        return x2,x_next,ae,ae2#x_prev, x_next, ae, ae2
```

```python
def translate_action(a):
    
    if a==0:
        action = np.array([0,1])
    elif a==1:
        action = np.array([0,-1])
    elif a==2:
        action = np.array([1,0])
    elif a==3:
        action = np.array([-1,0])
    return action
```

```python
def check_pos_valid(pos,action,graph):
    """ is this is valid move on the graph """
    newPos = cpos[0]+action[0],cpos[1]+action[1]
    n1,n2 = newPos
    s1,s2,_ = graph.shape
    if np.logical_and.reduce([n1>=0,
                              n2>=0,
                              n1<s1,
                              n2<s2]):
        valid = True
    else:
        valid = False
    #print(newPos,valid)
    return valid
```

```python
training_set = []
Vp_grid = Vp.reshape([12,12,144],order='F')

cpos = np.random.randint(0,12,size=2)

allX = []
allY = []
allA = []
action_list = []
for _ in range(2000):
    a = np.zeros(4)
    
    valid = False
    while not valid:
        a_ = np.random.randint(0,4)
        action = translate_action(a_)
        valid = check_pos_valid(cpos.copy(),action,Vp_grid)
        #time.sleep(0.1)

    
    a[a_] = 1
    action_list.append(a_)
    newPos = Vp_grid[cpos[0]+action[0],cpos[1]+action[1]]
    oldPos = Vp_grid[cpos[0],cpos[1]]
    
    cpos = np.array([cpos[0]+action[0],cpos[1]+action[1]])
    
    allX.append(oldPos)
    allA.append(a)
    allY.append(newPos)
    
allX = np.array(allX)
allY = np.array(allY)
allA = np.array(allA)
```

```python
criterion = nn.MSELoss()
model = Network()
optim = tf.optim.RMSprop(model.parameters(),lr=0.01)
```

```python
np.array(allY).shape
```

```python
#surely this isn't going to give you eigenvalues
epochs = 1500
hiddens = []
for e in range(epochs):
    running_loss = 0
    optim.zero_grad()

    x_prev,x_next,ae,ae2 = model(tf.tensor(allA,dtype=tf.float),
                             tf.tensor(allX,dtype=tf.float))
    
    hiddens.append(ae2)
    loss_pred = criterion(x_next, tf.tensor(allY,dtype=tf.float))
    loss_curr = criterion(x_prev,tf.tensor(allX,dtype=tf.float))
    
    if e%100==0:
        print(loss_pred.item())

    loss = loss_pred + .2*loss_curr#tf.sum(tf.abs(h))*0.0001
    
    #all_linear1_params = torch.cat([x.view(-1) for x in model.linear1.parameters()])
    #all_linear2_params = torch.cat([x.view(-1) for x in model.linear2.parameters()])
    #l1_regularization = lambda1 * torch.norm(all_linear1_params, 1)

    #This is where the model learns by backpropagating
    loss.backward()

    #And optimizes its weights here
    optim.step()

    running_loss += loss.item()
    #if e%100==0:
    #    print(loss.item())
#print("\nTraining Time (in minutes) =",(time()-time0)/60)
```
