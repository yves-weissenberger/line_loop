```python
class gridworld(object):
    """ Stupid gridworld environment class. Assume that is 4 connected"""
    def __init__(self,n_x,n_obj=None,cLoc=[0,0],rew_loc=[1,3]):
        if n_obj is None:
            self.n_obj = n_x
        else:
            self.n_obj = n_obj
        self.n_x = n_x
        #self.grid_obs = np.arange(0,n_x**2).reshape([n_x,n_x])
        #self.n_obj = np.unique(self.grid_obs)
        #self.grid_obs = np.vstack(n_x*[np.arange(0,n_x)])#np.random.randint(0,self.n_obj,size=n_x**2).reshape([n_x,n_x])
        self.grid_obs = np.random.randint(0,self.n_obj,size=n_x**2).reshape([n_x,n_x])
        self.n_obj = len(np.unique(self.grid_obs))
        self.grid_cost = np.ones([n_x,n_x])*-.01
        self.grid_cost[rew_loc[0],rew_loc[1]] = 1.
        #Current location
        self.cLoc = cLoc
        self.rew_loc = rew_loc
        
    def new_cLoc(self):
        self.cLoc = np.random.randint(0,self.n_x,size=2)
    def move(self,a):
        """ a is the action. [1,2,3,4]=['U','D','L','R']"""
        
        a = self._cvAct(a)
        
        self.cLoc = np.clip(self.cLoc +a,0,self.n_x-1)
        #print(self.cLoc)
        obs = self.grid_obs[self.cLoc[0],self.cLoc[1]]
        cost = self.grid_cost[self.cLoc[0],self.cLoc[1]]
        return self._toOneHot(obs), cost

    def get_obs(self,toOneHot=True):
        obs = self.grid_obs[self.cLoc[0],self.cLoc[1]]
        if toOneHot:
            return self._toOneHot(obs)
        else:
            return obs
    def _toOneHot(self,obs):
        tmp = np.zeros(self.n_obj)
        tmp[obs] = 1
        return tmp
    def _cvAct(self,a):
        if a==1:
            return np.array([-1,0])
        if a==2:
            return np.array([1,0])
        if a==3:
            return np.array([0,-1])
        if a==4:
            return np.array([0,1])
```

```python
class TEM(nn.Module):
    
    def __init__(self,nP,nG,n_classes):
        super(TEM, self).__init__()

        self.nP = nP
        self.nG = nG
        self.memW = .5

        
        ndim_embed = 5
        self.nTile_g = int(nP/nG) 
        self.nTile_mb = int(nP/ndim_embed)

        #for the inferenge model
        self.g_func = nn.Linear(nG+4+10,nG)
        self.lrlu = nn.LeakyReLU()
        self.rlu = nn.ReLU()
        self.converter = torch.from_numpy(np.array(list(set(permutations([1]*2 + [0]*3)))).T).type(torch.float32)
        
        #For the generative model
        #self.genG = nn.Linear(self.nG+4,self.nG)
        self.genG = net = nn.Sequential(
                                      nn.Linear(self.nG+4, 100),
                                      nn.LeakyReLU(),
                                      nn.Linear(100, self.nG),
                                      )
        self.toOut = nn.Linear(nP,n_classes)
        self.softmax = nn.Softmax(0)


    def forward(self,  obs, action,prev_g,prev_genG,MEM):
        ###########################
        #Implements inference model
        ###########################
        twoHot = torch.matmul(self.converter,obs)
        #twoHot_backup = twoHot.clone()

        x2 = torch.relu(twoHot - torch.mean(twoHot))
        x2 = x2/torch.norm(x2)
        x2_hippo = x2.repeat([self.nTile_mb])
        #x2_hippo = torch.matmul(x2_hippo,MEM)
        
        g_mu = torch.relu(self.g_func(torch.cat([prev_g,action,obs])))
        gdash_mu = g_mu.repeat([self.nTile_g])

        p_mu = self.lrlu(x2_hippo*gdash_mu)
        
        MEM_new = .999*MEM + self.memW*torch.ger(p_mu,p_mu)
        
        ############################
        #Implements generative model
        ############################
        
        gGen = self.rlu(self.genG(torch.cat([prev_g,action])))
        #genG_mem = gGen.repeat([self.nTile_g])
        
        pGen = torch.matmul(gdash_mu,MEM)
        
        
        output = self.softmax(self.toOut(pGen))
        return g_mu,p_mu,MEM_new,gGen,pGen,output

    def sample(self,prev_g,action,MEM):
        """ Sample based on posterior distribution """
        
        gGen = self.rlu(self.genG(torch.cat([prev_g,action])))
        pGen = torch.matmul(gGen.repeat([self.nTile_g]),MEM)
        return gGen,pGen
        

```

```python
def run_minibatch(gw,MEM,g_mu,hasvisited):
    allPred = []
    allObs = []
    diffS = []
    #g_mu = torch.zeros(tem.nG)
    gGen = torch.zeros(tem.nG)

    started = False
    for ijk in range(BTT):

        move = np.random.randint(4)+1
        obs = gw.move(move)[0]
        move_onehot = np.zeros(4)
        move_onehot[move-1] = 1

        g_mu,p_mu,MEM,gGen,pGen,output = tem.forward(obs=torch.from_numpy(obs).type(torch.float32),
                                                         action=torch.from_numpy(move_onehot).type(torch.float32),
                                                         prev_g=g_mu,
                                                         MEM=MEM,
                                                         prev_genG=gGen)
        
        diffS.append(np.array(g_mu.clone().data) - np.array(gGen.clone().data))

        allObs.append(np.argmax(obs))
        allPred.append(np.argmax(output.clone().detach().numpy()))
        first_time = hasvisited[gw.cLoc[0],gw.cLoc[1]]==0
        hasvisited[gw.cLoc[0],gw.cLoc[1]] += 1

        
        if not first_time:
            if not started:
                #print(1)
                loss = (l1(output.reshape(1,-1),torch.LongTensor([int(np.where(obs)[0])])) + 
                        l2(g_mu,gGen) + 
                        l2(p_mu,pGen))
            else:
                loss += (l1(output.reshape(1,-1),torch.LongTensor([int(np.where(obs)[0])])) + 
                                l2(g_mu,gGen) + 
                                l2(p_mu,pGen))

    loss.backward(retain_graph=True)
    loss.detach()
    for p in tem.parameters():
        p.data.add_(-learning_rate, p.grad.data)
    tem.zero_grad()
    return allPred, allObs, MEM.detach().numpy(), g_mu.detach().numpy(), diffS
```
