```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stt
```

```python
def get_pf(sz=200,xy=None,C=None):
    if xy==None:
        xM,yM = np.random.randint(0,sz,2)
    else:
        xM,yM = xy
    if C==None:
        C = np.eye(2)*np.random.randint(10,500,2)/2
    pF = np.zeros([sz,sz])
    mn = stt.multivariate_normal([xM,yM],C)
    a,b = np.meshgrid(np.arange(sz),np.arange(sz))
    pF = mn.pdf(np.vstack([a.flatten(),b.flatten()]).T).reshape(sz,sz)*10
    #for x in np.arange(sz):
    #    for y in np.arange(sz):
    #        pF[x,y] = multinorm(np.array([x,y]),np.array([xM,yM]),C)
    return pF
```

```python
place_fields = []
for x_ in range(1,20):
    for y_ in range(1,20):
        
        xy = [x_*20,y_*20]
        out = get_pf(xy=xy,C=20)
        place_fields.append(out)
place_fields = np.array(place_fields)
```

```python
l1 = []
l2 = []
for _ in range(2000):

    ix1 = np.random.randint(0,200,2)#np.array([10,20])
    ix2 = np.random.randint(0,200,2)
    l1.append(np.sum(np.abs(place_fields[:,ix1[0],ix1[1]] - place_fields[:,ix2[0],ix2[1]])))
    l2.append(np.sqrt(np.sum((ix1 - ix2)**2)))
```

```python
np.corrcoef(l1,l2)
```
