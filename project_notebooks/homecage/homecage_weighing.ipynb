```python
df = pd.read_excel("/Users/yves/Downloads/mouse_weights_homecage (3).xlsx")

only_after = 0; current = 0
data_manual = []
data_HC = []
for i,row in df.iterrows():
    if not pd.isnull(row['HC-1']):
        current+=1 
        if current>=only_after:
            data_manual.extend([row.values[i] for i in range(1,5)])
            data_HC.extend([row['HC-'+str(i)] for i in range(1,5)])
    
data_manual = np.array(data_manual)
data_HC = np.array(data_HC)

baseline_weights = np.array([float(i) for i in df.iloc[0].values[1:5]])
```

```python
dm_norm = (data_manual.reshape(-1,4)/baseline_weights[None]).flatten()
dhc_norm = (data_HC.reshape(-1,4)/baseline_weights[None]).flatten()

```

```python
dm_norm = (data_manual.reshape(-1,4)/baseline_weights[None]).flatten()
dhc_norm = (data_HC.reshape(-1,4)/baseline_weights[None]).flatten()
```

```python
clrs_plot = clrs*int(len(data_manual)/4.)
```

```python
plt.figure(figsize=(14,4))
plt.subplot(1,3,1)
min_ = np.min(np.concatenate([data_HC,data_manual]))
max_ = np.max(np.concatenate([data_HC,data_manual]))
plt.scatter(data_manual,data_HC,c=clrs_plot,s=96,alpha=.8)
plt.plot([min_,max_],[min_,max_],color='k',dashes=(7,7))
plt.xlabel("Manuel weights (g)")
plt.ylabel("homecage weights (g)")


plt.subplot(1,3,2)
min_ = np.min(np.concatenate([dhc_norm,dm_norm]))
max_ = np.max(np.concatenate([dhc_norm,dm_norm]))
plt.scatter(dm_norm,dhc_norm,c=clrs_plot,s=96,alpha=.8)
plt.axvline(x=.85,color='r',linewidth=3,alpha=.3)
plt.axhline(y=.85,color='r',linewidth=3,alpha=.3)

plt.plot([min_*.9,max_*1.1],[min_*.9,max_*1.1],color='k',dashes=(7,7))
plt.xlabel("Manuel weights \n(%baseline)")
plt.ylabel("homecage weights\n(%baseline)")



plt.subplot(1,3,3)
plt.title("Number of measurements {:.2f}".format(len(data_manual)))
seaborn.distplot(data_manual-data_HC,kde=0)
plt.xlabel("Error (Manual - homecage) (g)")
plt.ylabel("Number of measurements")




seaborn.despine()
```
