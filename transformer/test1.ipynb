```python
def layer_norm(x):
    m,v = np.mean(x),np.var(x)
    return (x-np.mean(x))/np.sqrt(v+1e-4)
```

```python
from itertools import combinations


def place_ones(size, count):
    for positions in combinations(range(size), count):
        p = [0] * size

        for i in positions:
            p[i] = 1

        yield p
```
